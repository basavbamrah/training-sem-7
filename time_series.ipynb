```python
X=[]
y=[]
for i in range(60,len(df)):
    X.append(df.iloc[i-60:i,5])
    y.append(df.iloc[i,5])

X=np.array(X)
y=np.array(y)
```

1. Recurrent Neural Network: in RNN the output of previous state is fed into input of next state
2. LSTM: Long Short Term Memory


```python
# No Standard Scalling, but minmax scaller

from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
# scaller wants 2d 
y=scaler.fit_transform(y.reshape((len(y),1)))
y=y.reshape((len(y),))

X=np.reshape(X,(X.shape[0],X.shape[1],1))
# convert data into 3D

import tensorflow as tf 

lstm=tf.keras.models.Sequential()
lstm.add(tf.keras.layers.LSTM(units=50,input_shape=(X.shape[1],1),return_sequences=True)) # return sequences
lstm.add(tf.keras.layers.LSTM(units=50))

lstm.add(tf.keras.layers.Dense(units=1))





lstm.compile(optimizer='adam',loss='mse',metrics=['mse'])
lstm.fit(X,y,epochs=10,batch_size=32)
```

```python
y_pred=lstm.predict(X)

y_pred=scaler.inverse_transform(y_pred)
y_pred
```
