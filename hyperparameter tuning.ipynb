```python
import pandas as pd
import numpy as np
df=pd.read_csv(r'./insurance.csv')
```

```python
x = df.iloc[:, :-1]
y = df['charges']
```

```python

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x['smoker'] = le.fit_transform(x['smoker'])
lo = LabelEncoder()
x['gender'] = lo.fit_transform(x['gender'])
x

# One Hot Encoder ----> "Non-ordinal variable"
ct = ColumnTransformer(transformers=[('encode', OneHotEncoder(), [5])], remainder='passthrough')
x = ct.fit_transform(x)
```

```python
# scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor as LR
regressor=LR()
```

```python
params={'min_samples_split':[2,6,10,20,50],
'min_samples_leaf':[2,6,10,20,50],
'splitter':['best','random'],
'max_leaf_nodes':[2,6,20,50,100]}


from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(regressor,params,scoring='r2',cv=5,verbose=3,n_jobs=3)
grid.fit(x_train,y_train)

# verbose is parameter which shows the 
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
regressor=LR(max_leaf_nodes=20,min_samples_leaf=6,min_samples_split =50,splitter='best')
regressor.fit(x_train,y_train)
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```
